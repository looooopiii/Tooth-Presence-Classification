```python
obj_path = "/local/scratch/datasets/Medical/TeethSeg/3DTeethLand_challenge_train_test_split/upper/015DTXR8/015DTXR8_upper.obj"
```

```python
import bpy
import sys
import os
```

```python
# Clear everything
bpy.ops.wm.read_factory_settings(use_empty=True)

# Load your 3D model (replace with your actual path)
model_path = obj_path
bpy.ops.import_scene.obj(filepath=model_path)

# Set render settings
output_dir = "D:/Studies/UZH-M.Sc/Master's Project/Tooth-Presence-Classification/output images"
bpy.context.scene.render.image_settings.file_format = 'PNG'
bpy.context.scene.render.resolution_x = 1024
bpy.context.scene.render.resolution_y = 1024
```

```python
# Set camera
def setup_camera(location, rotation_euler, name="Camera"):
    bpy.ops.object.camera_add(location=location, rotation=rotation_euler)
    cam = bpy.context.object
    cam.name = name
    bpy.context.scene.camera = cam
    return cam

# Set light
bpy.ops.object.light_add(type='SUN', location=(10, 10, 10))

# Views
views = {
    "top":    ((0, 0, 5),  (0, 0, 0)),
    "front":  ((0, -5, 0), (1.5708, 0, 0)),
    "side":   ((5, 0, 0),  (1.5708, 0, 1.5708)),
}

for view_name, (loc, rot) in views.items():
    setup_camera(loc, rot, name=f"Camera_{view_name}")
    bpy.context.scene.render.filepath = os.path.join(output_dir, f"{view_name}.png")
    bpy.ops.render.render(write_still=True)
    bpy.data.objects.remove(bpy.context.scene.camera)

print("Rendering complete.")
```
